```python
import numpy as np
import pandas as pd
import matplotlib as plt
import plotly.graph_objs as go
import yfinance as yf
```

### Canadian Imperial Bank

```python
data = yf.download(tickers = "CM", period = "5y", interval = "3mo")
```

```python
data
```

```python
#declare figure
fig = go.Figure()

#Candlestick
fig.add_trace(go.Candlestick(x = data.index, 
                             open = data["Open"],
                             high = data["High"], 
                             low= data["Low"], 
                             close = data["Close"]))
#Titles
fig.update_layout(title = "Canadian Imperial Bank live share price evolution",
                 yaxis_title = "Stock price (CAN per Shares)")

#X-axes

fig.update_xaxes(rangeslider_visible = True, rangeselector = dict(
    buttons = list([
        dict(count=15 ,label = "5y", step="year", stepmode = "todate")])))


fig.show()
```
