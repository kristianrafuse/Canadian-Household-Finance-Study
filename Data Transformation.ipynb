# Transforming the Data collected in a clean table

```python
# Dependencies and Setup
import pandas as pd

# Study data files
data_path = "quarterly.csv"

# Read the data 
#Due to errors on row 8 (which had no value assigned), we had to skip this specific row in order to read the file
data_original = pd.read_csv(data_path, encoding='utf-8', skiprows=8)

# Display the data table for preview
data_original.head()
```

```python
#Drop rows that are not relevant
data_df = data_original.drop([0, 1, 3, 30, 31, 32, 33, 34])

#Reset index
data_df = data_df.reset_index(drop=True)

#Rename main column
data_df = data_df.rename(columns={"Geography":"Categories"})

#Set new index as categories
data_df = data_df.set_index("Categories")

#Rename appropriate rows
data_df = data_df.rename(index={"Income, consumption and savings": "Year Quartile"})

#Display the data table
data_df
```

# Dividing the Data per Province

Calculating values per year and the great total

CANADA

```python
#Select olny the columns with values for Canada
canada_data = data_df.loc[:, ['Canada 1', 
                              'Canada 1.1', 
                              'Canada 1.2', 
                              'Canada 1.3', 
                              'Canada 1.4', 
                              'Canada 1.5', 
                              'Canada 1.6', 
                              'Canada 1.7', 
                              'Canada 1.8',
                              'Canada 1.9', 
                              'Canada 1.10']]


canada_data.head()
```

```python
#Calculate the yearly values for every category
CA_year = canada_data.drop(["Year Quartile"])

# Replace commas with empty strings in the columns to be summed
CA_year = CA_year.replace(',', '', regex=True)

# Convert values in DataFrame to numeric: https://pandas.pydata.org/docs/reference/api/pandas.to_numeric.html
CA_year = CA_year.apply(pd.to_numeric, errors='coerce', downcast="integer")

# Sum the values for each row in the specified columns
sum_col1 = CA_year.iloc[:, 0:4].sum(axis=1)
sum_col2 = CA_year.iloc[:, 4:8].sum(axis=1)
sum_col3 = CA_year.iloc[:, 8:11].sum(axis=1)

# Create a new dataframe with the summed values
canada_peryear = pd.DataFrame({'Canada 2020': sum_col1, 'Canada 2021': sum_col2, 'Canada 2022': sum_col3})

# Add a new column to the dataframe with the total sum of each row
canada_peryear['Total'] = canada_peryear.sum(axis=1)

# Format the numbers as currency: https://stackabuse.com/format-number-as-currency-string-in-python/
def format_currency(num):
    return "${:,.2f}".format(num)

# Apply the function to the dataframe
canada_peryear = canada_peryear.applymap(format_currency)

# Display the resulting dataframe
canada_peryear
```
