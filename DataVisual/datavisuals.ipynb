```python
import pandas as pd
import matplotlib.pyplot as plt

data = {'year': [2010, 2011, 2012, 2013, 2014, 2015, 2016],
        'card_debt': [100, 110, 100, 120, 130, 140, 150],
        'expenses': [110, 120, 110, 140, 160, 170, 180],
        'savings': [75, 65, 50, 40, 50, 35, 40],
        'housing_prices' : [70, 80, 90, 120, 150, 180, 200]}
        
df = pd.DataFrame(data)

df.plot(x='year', y= ['card_debt', 'expenses', 'housing_prices', 'savings'],
        kind='area', stacked=False, color=['red', 'blue', 'yellow', 'green']
       )
plt.xlabel('Year')
plt.ylabel('Finances')
plt.title('Canadian Finances')
```

```python
data = {'Year': [2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021],
        'Average Debt': [20000, 22000, 24000, 26000, 28000, 30000, 32000, 34000, 36000, 38000]}

df = pd.DataFrame(data)

df.plot.bar(x='Year', y='Average Debt', rot=0)

plt.xlabel('Year')
plt.ylabel('Average Debt ($)')
plt.title('Average Canadian Household Debt (2012-2021)')

plt.show()
```

```python
data = {'Year': [2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021],
        'Number of Credit Cards': [50000000, 55000000, 60000000, 65000000, 70000000, 75000000, 80000000, 85000000, 90000000, 95000000]}
df = pd.DataFrame(data)

df.plot.line(x='Year', y='Number of Credit Cards', marker='o')

plt.xlabel('Year')
plt.ylabel('Number of Credit Cards')
plt.title('Number of Credit Cards in Use in Canada (2012-2021)')

plt.show()
```
