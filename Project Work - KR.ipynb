```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#csv file path
quarterly = pd.read_csv("data/quarterly distributions.csv")
quarterly = quarterly.set_index('Income, consumption and savings')
```

```python
#read_csv, initial data organizing, tidying.
annually = pd.read_csv("data/annual distributions.csv")
annually = pd.DataFrame(annually)
annually = annually.set_index('Income, consumption and savings')
annually = annually.rename_axis(' ')
```

```python
#transpose csv to set-up for creating line graphs
annually_transposed = annually.transpose()
```

```python
#set the datatype of csv and remove special characters
annually_transposed = annually_transposed.replace(',', '', regex=True).astype(float)
annually_transposed
```

```python
#creating columns as a Percent of Income
annually_transposed['% Clothing and Footwear'] = 100* (annually_transposed['Clothing and footwear']/annually_transposed['Adjusted household disposable income'])
```

```python
#charting as a percent of income over 10 years
annually_transposed['% Clothing and Footwear'].plot(linewidth=6, kind='line')
plt.title('Percent of Income Spent on Clothing and Footwear')
plt.xlabel('Year')
plt.ylabel('Percent of Income')
plt.savefig("graphs/Percent of Income Spent on Clothing and Footwear")
```

```python
#setup the quarterly data for analysis
quarterly = quarterly.replace(',', '', regex=True).astype(float)
#get average month from last quater of 2022
quarterly['Monthly'] = quarterly['Q4 2022']/3
#transpose to setup for chart
quarterly_T = quarterly.T
```

```python
# Slice the DataFrame to include only the columns you want
quarterly_T_select = quarterly_T.loc['Monthly':, ['Food and non-alcoholic beverages', 'Alcoholic beverages and tobacco',
       'Clothing and footwear',
       'Housing, water, electricity, gas and other fuels',
       'Furnishings, household equipment and other goods and services related to the dwelling and property',
       'Health', 'Transport', 'Communications', 'Recreation and culture',
       'Education', 'Food, beverage and accommodation services',
       'Insurance and financial services', 'Miscellaneous goods and services']]

# Rename the column you want to change
quarterly_T_select = quarterly_T_select.rename(columns={'Furnishings, household equipment and other goods and services related to the dwelling and property': 'Furnishings and Household equipment'})

# Transpose the DataFrame so that the columns become rows
quarterly_T_select = quarterly_T_select.T

# Calculate the total of each row and create a new column with the total
quarterly_T_select['total1'] = quarterly_T_select.sum(axis=1)

# Create the pie chart with both percentage and actual values
fig, ax = plt.subplots(figsize=(20, 20))
quarterly_T_select.plot.pie(fontsize=12, y='total1', ax=ax, legend=False, autopct=lambda x: '{:.1f}% (${:.0f})'.format(x, x * quarterly_T_select['total1'].sum() / 100))
plt.title('Average Canadian Monthly Spending')

ax.set_xlabel('')
ax.set_ylabel('')
plt.savefig('graphs/Average Canadian Monthly Spending')
plt.show()
```

```python
#reading in new csv
debt = pd.read_csv("data/creditliabilities.csv")
```

```python
#removing every column that had "symbol" as the header and no data

symbol_columns = [col for col in debt.columns if 'Symbol' in col]

debt.drop(symbol_columns, axis=1, inplace=True)
```

```python
#selecting the credit card debt to analyze
cc_debt = pd.DataFrame(debt.iloc[4]).transpose()
```

```python
#setting the index so that i'm running commands below on just the data, now row headers.
cc_debt = cc_debt.set_index("Credit liabilities of households 1 2 3")
```

```python
cc_debt = cc_debt.replace(',', '', regex=True).astype(float)

cc_debt = cc_debt.T
```

```python
cc_debt.plot(linewidth=12, figsize=(20, 20), color='darkred')

plt.legend(loc='upper left', labels=['Household Credit Card Debt'])
plt.title('Total Canadian Household Credit Card Debt', fontsize=16)
plt.xlabel('January 2017 to Decemeber 2022')
plt.ylabel('Total Household Credit Card Debt x 1 000 000')
plt.savefig('graphs/Total Canadian Credit Card Debt')
```
